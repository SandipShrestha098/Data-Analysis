* 8 Fedu - father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)

* 16 schoolsup - extra educational support (binary: yes or no)

```python
def outlier_detection(data,columns):
  outliers = [] # list which will hold the outliers
  outliers_dict = {}
  outliers_count = 0

  for i in columns:
    q1, q3= np.percentile(data[i], [25,75]) # find mean of a particular column
    iqr = q3-q1

    lower_whisker = q1-1.5*iqr # lower bound of data
    upper_whisker = q3+1.5*iqr # upper bound of data

    # data outside of lower or upper whisker are considered as outliers

    print(f''' 
        column : {i}
        lower_whisker : {lower_whisker}
        upper_whisker :{upper_whisker}''')
    
    for x in data[i]:
      if (x < lower_whisker or x > upper_whisker):
        outliers.append(x) # adding outlies as list
        outliers_count += 1 # increasing the outliers count by 1 each time a new outlier is found
    
    print(f'''
        Total outliers : {len(outliers)}
        Outliers : {outliers}
        ''')
    
    outliers_dict.update({i:outliers_count}) # updating outliers dictionary with column as key and total outliers as its value
    outliers.clear() # clearing the outliers list
    outliers_count = 0 # resetting outliers count to 0 so that during next loop count starts from 0
  return outliers_dict
```

```python
int_data = cleaned_data.select_dtypes(int) # slicing only datas having integer values
int_data
```

```python
outlier_detection(data,int_data.columns.drop('failures'))
```

```python
def outlier_handeling(data,columns):
   for i in columns:
     q1, q3= np.percentile(data[i], [25,75]) # first and third quarter
     iqr = q3-q1 # inter quartile range

     lower_whisker = q1-1.5*iqr #lower bound
     upper_whisker = q3+1.5*iqr # upper bound
    
     for x in range(len(data)): # changes outliers to upper or lower whisker according to the closest value
       if data.loc[x, i] < lower_whisker: 
         data.loc[x, i] = lower_whisker
       if data.loc[x,i] > upper_whisker:
         data.loc[x, i] = upper_whisker
```

```python
copy_data = data.copy()  # making the copy of the data
outlier_handeling(copy_data,int_data.columns.drop('failures')) # removing outliers
```

## Encoding object type datas 

```python
encoded_data = pd.get_dummies(data =copy_data, columns = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob','reason', 'guardian', 'schoolsup', 'famsup', 'paid',
                                                          'activities','nursery', 'higher', 'internet', 'romantic'], drop_first = True ) # encoding object type data and droping the first column
```

```python
encoded_data.columns # columns after encoding
```

```python
encoded_data.shape # shape of encoded data
```

```python
X = encoded_data.drop('G3', axis = 1) # features for model
y = encoded_data['G3'] # target for model
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state= 60) # splitting train and test values
```

```python
print(x_train.shape, x_test.shape,y_train.shape,y_test.shape) # shape of train and test data
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() # creating a model
model.fit(x_train,y_train) # fitting model with train values
y_pred = model.predict(x_test) # predicting the test values
```

```python
from sklearn.metrics import mean_squared_error
import math  # importing maths for square root
MSE = mean_squared_error(y_test, y_pred) # mean squared error
RMSE = math.sqrt(MSE) # square root of mean squared error
print(f'''Mean Squared Error : {MSE}
Root Mean Squared Error : {RMSE}''')
```

The average of the squares of the differences between the predicted values and the true values is called 'Mean Squared Error(MSE)' and the linearRegression model have its value as 0.907789899378951. Similarly, the square root of the MSE is known as 'Root Mean Squared Error(RMSE)' and its value in LinearRegression is 0.9527800897263498 . The closer the value of MSE and RMSE to 0 better the model is.

```python
residual = y_test - y_pred
sns.histplot(residual, kde = True)
```

residual should have almost normal distribution

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print(f'''Mean Squared Error : {MSE}
Root Mean Squared Error : {RMSE}''')
```

The MSE and RMSE of Logistic Regression model is more far away from 0 so, the LinearRegression model is better in this case.

```python
residual = y_test - y_pred
sns.histplot(residual, kde = True)
```

Residual of logisticRegression is more normally distributed than linear regression.
